```python
import pandas as pd

# 1. RAW DATA: Typical "Wide" Vital Signs from a Rave eCRF
raw_vs_data = {
    'Subject': ['101-001', '101-002'],
    'Site': ['BER01', 'BER01'],
    'Visit': ['Screening', 'Screening'],
    'VSDAT': ['2026-01-10', '2026-01-12'],
    'WEIGHT': [75.5, 62.0],
    'HEIGHT': [180.0, 165.0],
    'SYSBP': [120, 118],
    'DIABP': [80, 78]
}

df_vs_wide = pd.DataFrame(raw_vs_data)

def map_to_sdtm_vs(df):
    """
    Transforms Wide Rave Vital Signs into Vertical (Long) SDTM VS Domain.
    """
    # List of parameters to "Melt" (Rotate)
    parameters = ['WEIGHT', 'HEIGHT', 'SYSBP', 'DIABP']
    
    # We use pd.melt to turn columns into rows
    vs_long = pd.melt(df, 
                      id_vars=['Subject', 'Site', 'Visit', 'VSDAT'], 
                      value_vars=parameters,
                      var_name='VSTESTCD', 
                      value_name='VSORRES')
    
    # Add SDTM Required Variables
    vs_long['STUDYID'] = 'BRAIN-TUMOR-2026'
    vs_long['DOMAIN'] = 'VS'
    vs_long['USUBJID'] = vs_long['STUDYID'] + "-" + vs_long['Site'] + "-" + vs_long['Subject']
    
    # VSTEST: Map the Code to the Full Name
    test_mapping = {
        'WEIGHT': 'Weight',
        'HEIGHT': 'Height',
        'SYSBP': 'Systolic Blood Pressure',
        'DIABP': 'Diastolic Blood Pressure'
    }
    vs_long['VSTEST'] = vs_long['VSTESTCD'].map(test_mapping)
    
    # VSORRESU: Add Units (Simplified for this demo)
    unit_mapping = {
        'WEIGHT': 'kg',
        'HEIGHT': 'cm',
        'SYSBP': 'mmHg',
        'DIABP': 'mmHg'
    }
    vs_long['VSORRESU'] = vs_long['VSTESTCD'].map(unit_mapping)
    
    # Date Format
    vs_long['VSDTC'] = pd.to_datetime(vs_long['VSDAT']).dt.strftime('%Y-%m-%d')
    
    return vs_long[['USUBJID', 'VSTESTCD', 'VSTEST', 'VSORRES', 'VSORRESU', 'VSDTC']]

# Execute
sdtm_vs_df = map_to_sdtm_vs(df_vs_wide)
print("\n--- CDISC SDTM VS Domain (Long Format) ---")
print(sdtm_vs_df.sort_values(by=['USUBJID', 'VSTESTCD']))
```
