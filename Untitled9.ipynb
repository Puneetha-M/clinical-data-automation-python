```python
import pandas as pd
import numpy as np

# 1. SIMULATED DATA: Raw Adverse Event Export from Medidata Rave
# This mimics what you would see in a 'Standard' AE log in Rave
raw_ae_data = {
    'Subject': ['101-001', '101-001', '101-002'],
    'Site': ['BER01', 'BER01', 'BER01'],
    'AETERM_RAW': ['Headache', 'Nausea', 'Dizziness'],
    'AESTDAT': ['2026-02-01', '2026-02-05', '2026-03-01'],
    'AEENDAT': ['2026-02-02', '', '2026-03-05'], # Empty string means 'Ongoing'
    'AESEV': ['Mild', 'Moderate', 'Severe'],
    'AESER': ['No', 'No', 'Yes'],
    'AEREL': ['Related', 'Not Related', 'Possibly Related']
}

df_ae_raw = pd.DataFrame(raw_ae_data)

def map_to_sdtm_ae(df):
    """
    Transforms Raw Rave AE data into CDISC SDTM AE Domain.
    """
    ae = pd.DataFrame()
    
    # Required SDTM Variables
    ae['STUDYID'] = 'BRAIN-TUMOR-2026'
    ae['DOMAIN'] = 'AE'
    ae['USUBJID'] = ae['STUDYID'] + "-" + df['Site'] + "-" + df['Subject']
    
    # AESEQ: Sequence number per subject
    ae['AESEQ'] = df.groupby('Subject').cumcount() + 1
    
    # AETERM: The reported term from Rave
    ae['AETERM'] = df['AETERM_RAW'].str.upper()
    
    # AEDECOD: In a real trial, you'd use a MedDRA API. Here we simulate the mapping.
    ae['AEDECOD'] = ae['AETERM'] # Dictionary coding simulation
    
    # Dates: Converting to ISO 8601
    ae['AESTDTC'] = pd.to_datetime(df['AESTDAT']).dt.strftime('%Y-%m-%d')
    ae['AEENDTC'] = pd.to_datetime(df['AEENDAT']).dt.strftime('%Y-%m-%d').replace('NaT', '')
    
    # AESEV: Severity (Controlled Terminology)
    ae['AESEV'] = df['AESEV'].str.upper()
    
    # AESER: Serious Event (Mapping 'Yes/No' to 'Y/N')
    ae['AESER'] = df['AESER'].map({'Yes': 'Y', 'No': 'N'})
    
    # AEREL: Causality/Relationship
    ae['AEREL'] = df['AEREL'].str.upper()
    
    # AEENRF: End Relative to Reference Period (Ongoing flag)
    ae['AEENRF'] = np.where(df['AEENDAT'] == '', 'ONGOING', '')
    
    return ae

# Execute AE Transformation
sdtm_ae_df = map_to_sdtm_ae(df_ae_raw)

print("--- CDISC SDTM AE Domain Generated ---")
print(sdtm_ae_df[['USUBJID', 'AETERM', 'AESTDTC', 'AESEV', 'AESER', 'AEENRF']])
```
