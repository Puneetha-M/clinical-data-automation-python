```python
import pandas as pd

# 1. RAW DATA
raw_data = {
    'Subject': ['101-001', '101-002', '101-003'],
    'Site': ['BER01', 'BER01', 'MUN02'],
    'BRTHDT': ['1985-05-12', '1990-11-20', '1978-02-15'],
    'SEX': ['Male', 'Female', 'Male'],
    'Informed_Consent_Date': ['2026-01-10', '2026-01-12', '2026-01-15']
}
df_raw = pd.DataFrame(raw_data)

def map_to_sdtm_dm(df):
    """
    Transforms Raw Rave Data into CDISC SDTM DM (Demographics) Domain.
    """
    # We initialize the DM dataframe using the same index as the input df
    dm = pd.DataFrame(index=df.index)
    
    # Define the study ID once
    study_id = 'BRAIN-TUMOR-2026'
    
    # Assigning
    dm['STUDYID'] = study_id
    dm['DOMAIN'] = 'DM'
    
    # USUBJID: Study-Site-Subject
    dm['USUBJID'] = study_id + "-" + df['Site'] + "-" + df['Subject']
    
    dm['SUBJID'] = df['Subject']
    
    # Dates
    dm['RFSTDTC'] = pd.to_datetime(df['Informed_Consent_Date']).dt.strftime('%Y-%m-%d')
    dm['BRTHDTC'] = pd.to_datetime(df['BRTHDT']).dt.strftime('%Y-%m-%d')
    
    # Mapping
    dm['SEX'] = df['SEX'].map({'Male': 'M', 'Female': 'F', 'Other': 'U'})
    dm['ARM'] = 'ACTIVE TREATMENT'
    
    return dm

# Execute and Print
sdtm_dm_df = map_to_sdtm_dm(df_raw)
print("--- CDISC SDTM DM Domain Generated ---")
print(sdtm_dm_df)
```
